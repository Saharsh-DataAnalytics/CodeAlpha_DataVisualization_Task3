```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import matplotlib.image as mpimg
```

```python
df = pd.read_csv('SuperstoreSales_Cleaned.csv')
```

```python
fig = plt.figure(figsize = (26,30))
fig.patch.set_facecolor("#F8F9FA")

gs = GridSpec(
    nrows=28,
    ncols=16,
    figure=fig
)

logo = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/Company_logo.png")

kpi = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/KPI.png")

sales = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/SalesbyCategory.png")

profit = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/ProfitbyCategory.png")

furniture = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/ProfitbySubCategory.png")

discount = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/DiscountProfitScatter.png")

region = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/ProfitbyRegion.png")

segment = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/ProfitbySegments.png") 

ship = mpimg.imread(r"C:\Users\ASUS\OneDrive/Documents/Data Analytics Notes/DA_Intern_Prep/CodeAlpha/Task3Visualization/Images/ProfitbyShipMode.png")



ax_logo = fig.add_subplot(gs[0:2,0:2])
ax_logo.imshow(logo)
_ = ax_logo.axis('off')

ax_title = fig.add_subplot(gs[0:2,2:12])

ax_title.text(
    0.02,
    0.5,
    'Superstore Profitability Analysis Dashboard',
    fontsize = 40,
    fontweight = 'bold',
    va = 'center',
    ha = 'left'
)
ax_title.axis('off')

ax_kpi = fig.add_subplot(gs[2:7,2:14])
ax_kpi.imshow(kpi)
ax_kpi.axis('off')


ax_sales = fig.add_subplot(gs[7:12,0:8])
ax_sales.imshow(sales)
ax_sales.axis('off')

ax_profit = fig.add_subplot(gs[7:12,8:16])
ax_profit.imshow(profit)
ax_profit.axis('off')

ax_furniture = fig.add_subplot(gs[12:17,0:8])
ax_furniture.imshow(furniture)
ax_furniture.axis('off')

ax_discount = fig.add_subplot(gs[12:17,8:16])
ax_discount.imshow(discount)
ax_discount.axis('off')

ax_region = fig.add_subplot(gs[17:23,0:8])
ax_region.imshow(region)
ax_region.axis('off')

ax_segment = fig.add_subplot(gs[17:23,8:16])
ax_segment.imshow(segment)
ax_segment.axis('off')

ax_ship = fig.add_subplot(gs[23:28,0:8])
ax_ship.imshow(ship)
ax_ship.axis('off')

plt.tight_layout(pad=2.0)

plt.savefig(
    "Superstore_Profitability_Dashboard.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
